```python
import time as t
import re
import random
class ATM:
    def __init__(self):
        self.user={}
        self.current_user=""
    def create_user(self):
        lst=[]
        name=input("Enter Your Username: ")
        if name not in self.user:
            lst.append(name)
            print("Minimum eight,at least 1 uppercase,1 lowercase,1 number & 1 special character:")
            pas=input("Enter Your Password: ")
            pattern="(?=.*[a-z])(?=.*[A-Z])(?=.*\d)(?=.*[@$!%*?&])[A-Za-z\d@$!%*?&]{8,}"
            r=re.match(pattern,pas)
            if r:
                lst.append(pas)
            amo=int(input("Initial Deposit Amount:"))
            lst.append(amo)
            self.user[lst[0]]=lst[1:]
            print("Successfully Created")
    
    def balance(self):
        u1=input("Enter your user name:")
        if u1 in self.user:
            pas=input("Enter your password:")
            if self.user[u1][0]==pas:
                self.current_user=u1
                print("Your Balance is ",self.user[u1][1])
            else:
                print("Password Is wrong")
                
        else:
            print("There is no user")
            b=input("Want to create account(Y/N):")
            if b=='Y':
                atm.create_user()
            else:
                return
    def withdrawl(self,q):
        if self.current_user!="":
            if self.user[self.current_user][1]>=q:
                otp=random.randint(100000,999999)
                print("The OTP: ",otp,"\n","Enter the OTP:")
                if otp==int(input()):
                    self.user[self.current_user][1]-=q
                    print("Collect your amount","\n","Your balance is",self.user[self.current_user][1])
            else:
                print("Insufficient Balance",self.user[self.current_user][1])
        else:
            print("Please login...")
            t.sleep(2)
            u1=input("Enter your user name:")
            if u1 in self.user:
                pas=input("Enter your password:")
                if self.user[u1][0]==pas:
                    self.current_user=u1
                    print("Successfully Logined")
    def deposit(self,q):
        if self.current_user!="":
            self.user[self.current_user][1]+=q
            print("Your Final Balance is",self.user[self.current_user][1])
        else:
            print("Please login...")
            t.sleep(2)
            u1=input("Enter your user name:")
            if u1 in self.user:
                pas=input("Enter your password:")
                if self.user[u1][0]==pas:
                    self.current_user=u1
                    print("Successfully Logined")
            else:
                print("Account not found")
                return
            
            
        
            
if __name__=="__main__":
    print("       Welcome to New ATM       ")
    atm=ATM()
    while True:
        print("What's your Choice","\n")
        print(" 1.Create Account")
        print(" 2.Check Balance")
        print(" 3.Withdrawl")
        print(" 4.Deposit")
        print(" 5.Exit")
        a=int(input("Please enter:"))
        if a==1:
            atm.create_user()
        elif a==2:
            atm.balance()
        elif a==3:
            q=int(input("Enter Amount to Withdrawl:"))
            atm.withdrawl(q)
        elif a==4:
            q=int(input("Enter Amount You are Depositing:"))
            atm.deposit(q)
        elif a==5:
            print("Thank You")
            t.sleep(3)
            break
        else:
            print("Sorry! Invalid Input")
            print("So....")
            t.sleep(5)
            print("We are leaving")
            break
        
```
